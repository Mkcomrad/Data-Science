```python
df=df2
```

```python
# Convert vote-related columns to numeric
df['EVM Votes'] = pd.to_numeric(df['EVM Votes'].str.replace(',', ''), errors='coerce')
df['Postal Votes'] = pd.to_numeric(df['Postal Votes'].str.replace(',', ''), errors='coerce')
df['Total Votes'] = pd.to_numeric(df['Total Votes'].str.replace(',', ''), errors='coerce')
df['Vote Share'] = pd.to_numeric(df['Vote Share'].str.rstrip('%'), errors='coerce')  # Remove '%' and convert to float

# Display the first few rows after conversion
print(df.head())
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Party', data=df, palette='viridis', order=df['Party'].value_counts().index)
plt.title('Distribution of Parties')
plt.xlabel('Party')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
top_states = df['State'].value_counts().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_states.index, y=top_states.values, palette='viridis')
plt.title('Top 10 States by Number of Candidates')
plt.xlabel('State')
plt.ylabel('Number of Candidates')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Total Votes'], bins=30, kde=True, color='purple')
plt.title('Total Votes Distribution')
plt.xlabel('Total Votes')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Party', y='Total Votes', data=df, palette='viridis')
plt.title('Total Votes by Party')
plt.xlabel('Party')
plt.ylabel('Total Votes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['EVM Votes'], df['Postal Votes'], alpha=0.7)
plt.title('EVM Votes vs Postal Votes')
plt.xlabel('EVM Votes')
plt.ylabel('Postal Votes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Party', y='Vote Share', data=df, palette='viridis')
plt.title('Vote Share by Party')
plt.xlabel('Party')
plt.ylabel('Vote Share (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
top_constituencies = df.nlargest(10, 'Total Votes')

plt.figure(figsize=(12, 6))
sns.barplot(x='PC Name', y='Total Votes', data=top_constituencies, palette='viridis')
plt.title('Top 10 Constituencies by Total Votes')
plt.xlabel('Constituency')
plt.ylabel('Total Votes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
